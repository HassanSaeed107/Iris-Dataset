```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

```

## Importing the Data and View First 5 rows of columns

```python
data = pd.read_csv('/content/bezdekIris.data')
data.head()
```

## Renaming the Columns names

```python
data = data.rename(columns={
    '5.1': 'SepalLength',
    '3.5': 'SepalWidth',
    '1.4': 'PetalLength',
    '0.2': 'PetalWidth',
    'Iris-setosa': 'Species'})
```

```python
data
```

## Information about Dataset

```python
data.info()
```

```python
data.isnull().sum()
```

## Removes the null values

```python
data.dropna()
```

## Statistics summary of Numeric data

```python
data.describe()
```

## **Pairplot:**

"We created a pairplot, which allowed us to visualize pairwise relationships between the features. By coloring the points based on species, we were able to observe distinct clusters for each species."

```python
sns.pairplot(data, hue='Species')
plt.title('Pair Plot of Iris DataSet')
plt.show()
```

## **Correlation Heatmap:**

"We also generated a correlation heatmap to visualize the correlations between different features. We found strong positive correlations between petal length and petal width, as well as between sepal length and petal length."

## Correlation heatmap

```python
plt.figure(figsize =(8,6))
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Map')
plt.show()
```

## **Outlier Detection:**

"Using box plots, we detected potential outliers in the dataset. However, given the nature of our data, we decided to keep these outliers as they may represent genuine variations in flower measurements."

## Outlier Detection (Box Plot)

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=data)
plt.title('Box plot of Iris data')
plt.show()
```

## **Data Distribution Analysis:**

"Finally, we analyzed the distribution of data using histograms. We observed that the distributions of sepal length and sepal width were approximately normal, while petal length and petal width exhibited bimodal distributions."

## Data Distribution Analysis (Histogram)

```python
data.hist(figsize=(10,8))
plt.suptitle("Histogram of Iris Dataset", x =0.5, y=1.02, ha='center')
plt.show()
```
